### Age

```python
plt.figure(figsize = (12,5))

df['Age'].value_counts().sort_index().plot()
plt.xticks(range(min(df['Age']), max(df['Age'])+1, 2))
plt.title('Age Vs Employee Count', color = 'black')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
sns.displot(kind = 'hist', data = df, x = 'Age', hue = 'Attrition')
plt.plot()
```

```python
sns.catplot(kind = 'box', data = df, x = 'Age')
```

```python
df['Age'].describe()
```

#### Conclusion:
- Employeees in age of 25 to 40 are more likely to leave their job.
- There are no outliers in the age data.



### DistanceFromHome

```python
plt.figure(figsize = (12,5))
sns.displot(kind = 'hist', data = df, x = 'DistanceFromHome', hue = 'Attrition', bins = 10)
plt.xticks(range(min(df['DistanceFromHome']), max(df['DistanceFromHome'])+1, 5))
plt.show()
```

```python
avg_distance_from_home = df['DistanceFromHome'].mean().round(2)
print('Average distance from home:', avg_distance_from_home)
```

```python
sns.catplot(kind = 'box', data = df, x = 'DistanceFromHome')
```

```python
df['DistanceFromHome'].describe()
```

#### Conclusion:
- Lesser the distance more number of employees are working. As a result the arrtition is high.
- Average distance from home is 9.2 unit.
- There is no outliers.
